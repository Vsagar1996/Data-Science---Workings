## Data Extraction with Pandas

#### Selecting data : loc(), iloc() methods

```python
# loc() is label based, so we have to specify rows and columns based on their index/row and column labels
# iloc() is integer position based, so we have to specify rows and columns by their integer position values (0 based integer position)
```

```python
df = pd.read_csv("players_20.csv")
pd.set_option("display.max_rows",20)  # here n is the total number of rows to be placed
df
```

Difference between loc and iloc methods

loc:

syntax: df.loc['Zero'] - for value
        df.loc[['Zero','two']] - for list
        df.loc['zero':'two'] - for slicing

iloc:

syntax: df.iloc[0] - for value
        df.iloc[[0,2]] - for list
        df.iloc[[0:2]] - for slicing

In loc method, the start and end elements of the slicing is included
But in iloc method, the start element is only included

### 4. Selecting with conditions

iloc cannot accept a boolean series, but only a boolean list.

We have to use the list() function to convert a series into a boolean list

## Filter a dataframe using the query() method

```python
# select players older than 34

df = pd.read_csv("players_20.csv")
df.set_index("short_name",inplace = True)
df.query("age>34")
```

```python
# write the equivalent boolean slicing for the above example

df[df["age"]>34]
```

```python
# select players older than 34 from italy

df.query("age>34 and nationality=='Italy'")
```

```python
# add a not operator to the first example

df.query("not(age>34)")
```

```python
# convert height to meters and select those with height above 1.8

df.query("(height_cm/100)>1.8")
```

```python
# select players that were born after 1990

# Convert 'dob' column to datetime
df['dob'] = pd.to_datetime(df['dob'], errors='coerce')  # 'coerce' turns invalid dates into NaT

# Extract the year
df['year'] = df['dob'].dt.year

# query

df.query("dob.dt.year > 1990")
```

```python
# backticks is used if the column name has spaces or special characters

dff = pd.DataFrame({"A C":[2,4,5,2],"B@1":["a","b","c","d"]})
dff.query(" `A C` == 2 and `B@1` == 'a' ")
```

### .apply() function

```python
# use numpy function and apply it to series

df['age'].apply(np.sqrt)  # only creates a copy of the original dataframe
```

```python
# create your own function and apply it to a dataframe

def convert(row):
    return row["height_cm"]/100

df.apply(convert, axis=1)
```

```python
def convert2(col_val):
    return (col_val)/100

df["height_cm"].apply(convert2)
```

```python
def classify(col_val):
    if col_val >= 180:
        return "Tall"
    elif col_val < 180 and col_val >=150:
        return "Medium"
    else:
        return "Short"

df["Height Category"] = df["height_cm"].apply(classify)
df
```

```python
# using lambda function and apply it to a dataframe

df["height_cm"].apply(lambda X: X/100)

# alternative method - df["height_cm"]/100
```

```python
# use lambda function to convert "long_name" series to upper case

df["long_name"].apply(lambda X: str(X).upper())

# alternate with str attribute  -  df["long_name"].str.upper()
```

```python
# use lambda and apply to get the year in "dob" series

df['dob'] = pd.to_datetime(df['dob'], errors='coerce')
df["dob"].apply(lambda X: X.year)
```

```python
# apply lambda function to dataframe in order to calculate bmi

df.apply(lambda X: X["weight_kg"] / ((X["height_cm"]/100)**2), axis=1)
```

### Make a copy of a dataframe with copy()

#### 1.1  deep = True

```python
# deep = True by default (modifications to the data or indices at original doesnot reflect in the copy and vice versa)

df_copy = df.copy()

# update value in original dataframe

df.loc["L. Messi","height_cm"] = 180

# value will updated in the original dataframe, but not in the copy and vice versa

df_copy
```

#### 1.2  deep = False

```python
# by keeping deep = False, any change in the original df will reflect in the copy and vice versa

df_copy1 = df.copy(deep=False)
df_copy1
```

```python
# update value in original dataframe - makes a shallow copy

df.iloc[1,6] = 200
```

```python
df_copy1
```

#### Simple assignment - Another way to make a shallow copy

```python
df_newcopy = df
```

```python
df.iloc[2,6] = 180

df_newcopy
```

## Reshaping and Pivoting Dataframes

### Pivot Tables - pivot(), pivot_table()

```python
# pivot() - reshape data based on column values (doesnot support data aggregation)

# syntax - df.pivot(index = "any_column",columns="any_column",values="any_column")
```

```python
# Example 1

data = {"foo":["one","one","one","two","two","two"],
        "bar":["A","B","C","A","B","C"],
        "baz":[1,2,3,4,5,6],
       "zoo":["x","y","z","q","w","t"]}
df = pd.DataFrame(data)
df
```

```python
df.pivot(index="foo",columns="bar",values="baz")
```

## pivot_table() - creates a spreadsheet style pivot table (supports data aggregation)

```python
# new dataset
df_sales = pd.read_excel("supermarket_sales.xlsx")
df_sales
```

```python
df_sales.select_dtypes(include=["number"])

#df_sales.select_dtypes("number").columns

#df_sales.select_dtypes("number").mean()
```

```python
# make a pivot table apply an aggregate function on all numeric columns

# Select only numeric columns (excluding datetime)
numeric_cols = df_sales.select_dtypes(include=["number"]).columns

# Create the pivot table using only numeric columns
df_sales.pivot_table(index="Gender", values=numeric_cols, aggfunc="sum")

#df_sales.pivot_table(index="Gender",aggfunc="sum") - showing error (datetime64 type does not support sum operations) on this tuitorial method
```

```python
# make a pivot table, apply an aggregate function on some selected columns

df_sales.pivot_table(index="Gender",values=["Quantity","Total"],aggfunc="sum")
```

```python
# make a pivot table thta shows how much male and fenmale spend in each product and quantity brought

df_sales.pivot_table(index="Gender",columns="Product line",values=["Quantity","Total"],aggfunc="sum")
```

```python
df_sales.pivot_table(index=["City","Gender"],columns="Product line",values="Total",aggfunc="sum")
```
