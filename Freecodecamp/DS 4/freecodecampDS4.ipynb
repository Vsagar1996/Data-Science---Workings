### The Split-Apply-Combine strategy

```python
# split data into separate groups

passenger_filter = df_cars["Vehicle_type"]=="Passenger"
car_filter = df_cars["Vehicle_type"]=="Car"
```

```python
# apply an operation

passenger_avg = df_cars[passenger_filter]["Sales_in_thousands"].mean()
car_avg = df_cars[car_filter]["Sales_in_thousands"].mean()
```

```python
# combine results into a dataframe

pd.DataFrame({"Vehicle_type":["Car","Passenger"],
             "Avg_Sales_in_thousands":[car_avg,passenger_avg]}).set_index("Vehicle_type")
```

## groupby() and agg() functions together

```python
# find the minimum and maximum values on each column

df_cars.groupby("Vehicle_type").agg(["min","max"])
```

```python
df_cars.groupby("Vehicle_type")[["Sales_in_thousands","Price_in_thousands","Engine_size","Horsepower","Fuel_capacity"]].agg(["sum","mean"])
```

```python
numerical_cols = df_cars.select_dtypes("number").columns
df_cars.groupby("Vehicle_type")[numerical_cols].agg(["sum","mean"])
```

```python
# set a name for aggregated column and find the min Engine size and max horsepower

df_cars.groupby("Vehicle_type").agg(min_engine_size=("Engine_size","min"),
                                    max_horsepower=("Horsepower","max"))
```

```python
# calculate the sum of sales_in_thousands and mean of price_in_thousands

df_cars.groupby("Manufacturer").agg(sum_sales=("Sales_in_thousands","sum"),mean_price=("Price_in_thousands","mean"))
```

```python
# Alternate way for the above example without using agg function

sum_sales = df_cars.groupby("Manufacturer").sum()["Sales_in_thousands"]
mean_price = df_cars.groupby("Manufacturer").mean(numeric_only=True)["Price_in_thousands"]
df_Manufacturer = pd.DataFrame([sum_sales,mean_price])
df_Manufacturer = df_Manufacturer.T
df_Manufacturer
```

```python
sum_sales = df_cars.groupby("Manufacturer")["Sales_in_thousands"].sum()
mean_price = df_cars.groupby("Manufacturer")["Price_in_thousands"].mean()
df_Manufacturer = pd.DataFrame([sum_sales,mean_price])
df_Manufacturer = df_Manufacturer.T
df_Manufacturer
```

## groupby with lambda function

```python
# get actual sales and price values

df_cars.groupby("Manufacturer").sum().agg(lambda X:X*1000)[["Sales_in_thousands","Price_in_thousands"]]
```

```python
# substract values from the mean of each group

df_cars.groupby("Manufacturer").apply(lambda X:X-X.mean(numeric_only=True))
```

```python
df_cars.groupby("Manufacturer").mean(numeric_only=True)
```

## filtering groups with filter()

```python
def filter_num(x):
    return x>52

filter_num(58)
```

```python
# filter manufacturers out based on sum() and return a subset

def filter_func(x):
    """ x is the dataset"""
    return (x["Sales_in_thousands"].sum() > 52)

df_filter = df_cars.groupby(["Manufacturer"]).filter(filter_func)
df_filter
```

```python
# # manufacturers with sum of sales more than 52

df_filter.groupby(["Manufacturer"])["Sales_in_thousands"].sum()
```

```python
# manufacturers with sum of sales less than 52

df_cars.groupby(["Manufacturer"]).sum()["Sales_in_thousands"].sort_values().head(6)
```

```python
df_removed = df_cars[df_cars["Manufacturer"].isin(["Porsche","Jaguar","Saab","Infiniti","Audi","BMW"])]
df_removed
```
