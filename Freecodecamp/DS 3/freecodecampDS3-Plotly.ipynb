```python
import plotly.express as px
import pandas as pd
```

```python
# dataset

df_population = pd.read_csv("population_total.csv")
df_population
```

```python
# dropping null values

df_population.dropna(inplace=True)
```

```python
# making a pivot table - reshape the dataframe and save

df_pivot= df_population.pivot(index="year",columns="country",values="population")
df_pivot
```

```python
# select some countries - simplifying the dataframe and overwrite

df_pivot = df_pivot[["United States","India","China","Indonesia","Brazil"]]
df_pivot
```

```python
df_sample = df_population[df_population["country"].isin(["United States","India","China","Indonesia","Brazil"])]
df_sample
```

```python
# reset the index and change it to a column and rename

df_reset1 = df_pivot.reset_index()
#df_reset1

#Melt the DataFrame for Plotly Express

df_melted1 = df_reset1.melt(id_vars='year', var_name='country', value_name='population')
#df_melted1

# Create the line plot with axis labels

fig = px.line(df_melted1, x='year', y='population', color='country',
             labels={'year': 'Years', 'population': 'Population', 'country': 'Country'},
             title='Population Growth Over Time')
fig.show()
```

```python
# adding specifuc colors to the lines

df_reset1 = df_pivot.reset_index()
#df_reset1

#Melt the DataFrame for Plotly Express

df_melted1 = df_reset1.melt(id_vars='year', var_name='country', value_name='population')
#df_melted1

# Create the line plot with axis labels

fig = px.line(df_melted1, x='year', y='population', color='country',
             labels={'year': 'Years', 'population': 'Population', 'country': 'Country'},
             color_discrete_sequence=['green','red','blue','brown','cyan'],
             title='Population Growth Over Time')
fig.show()
```

## Interactive Piechart

```python
# dataset

df_pivot_2015 = df_pivot[df_pivot.index.isin([2015])].T.T
df_pivot_2015
```

To plot a pie chart from your pivot DataFrame showing the population distribution across countries in 2015 using Plotly Express, you’ll first need to reshape the data so it has two columns: country and population.

```python
df_pivot_2015_reset = df_pivot_2015.reset_index()

#df_pivot_2015_reset.columns = ['country', 'population'] # to rename the columns and also the remove the name above index
df_pivot_2015_reset
```

```python
df_pivot_2015_melt = df_pivot_2015_reset.melt(id_vars = "year",var_name = "country",value_name = "population")
df_pivot_2015_melt
```

```python
# Plot pie chart

fig = px.pie(df_pivot_2015_melt, names='country', values='population',
             title='Population Distribution by Country in 2015',
             labels={'country': 'Country', 'population': 'Population'},
             hole=0.3)  # optional: makes it a donut chart
fig.show()
```

```python
# Create the pivot DataFrame
df_pie = pd.DataFrame({
    'United States': [3.208783e+08],
    'India': [1.310152e+09],
    'China': [1.406848e+09],
    'Indonesia': [2.583833e+08],
    'Brazil': [2.044718e+08]})

# Reshape the data

df_pie_melted = df_pie.melt(var_name='country', value_name='population')

# Plot pie chart

fig = px.pie(df_pie_melted, names='country', values='population',
             title='Population Distribution by Country in 2015',
             labels={'country': 'Country', 'population': 'Population'})

fig.update_traces(textinfo='value+percent+label', textposition = 'outside', hovertemplate='%{label}: %{value:,}')
fig.show()
```

## Interactive boxplot

```python
# dataset

df_UnitedStates = df_pivot[["United States"]]
df_UnitedStates
```

```python
df_UnitedStates_reset = df_UnitedStates.reset_index()
df_UnitedStates_reset.columns = ["year","population"]
df_UnitedStates_reset
```

```python
# Create boxplot

fig = px.box(df_UnitedStates_reset, y='population',
             points='all',  # show all data points
             title='Population Distribution of United States (1955–2020)',
             labels={'population': 'Population'})

fig.update_traces(marker=dict(size=6, color='red'))
fig.show()
```

## Interactive Histograms

```python
# Dataset

df_hist = df_pivot[["United States","Indonesia"]]
df_hist
```

```python
# Reshape the data

df_hist_reset = df_hist.reset_index().rename(columns={'index':'year'})
df_hist_reset

# Melt the DataFrame

df_hist_melted = df_hist_reset.melt(id_vars='year', var_name='country', value_name='population')
df_hist_melted

# Create histogram

fig = px.histogram(df_hist_melted, x='population',color='country',
                   nbins=10,  # adjust number of bins
                   title='Population Distribution Histogram (United States vs Indonesia)',
                   labels={'population': 'Population', 'country': 'Country'},
                   #marginal='box',  # adds boxplot on side
                  )

fig.update_layout(bargap=0.1)
fig.show()
```

## Interactive Scatterplot

```python
# dataset

df_pivot
```

To create an interactive scatter plot using Plotly Express from your pivot-style DataFrame (countries as columns, years as index), you'll first need to reshape the data into a long-form format. This allows each country-year-population trio to be plotted as a point.

```python
# reshaping the dataframe

df_melted4 = df_pivot.reset_index().melt(id_vars='year', var_name='country', value_name='population')
df_melted4
```

```python
# to plot the scatter plot

fig = px.scatter(df_melted4, x='year', y='population', color='country',
                 title='Population Growth Scatter Plot (1955–2020)',
                 labels={'year': 'Year', 'population': 'Population', 'country': 'Country'},
                 hover_data={'year': False, 'population': ':.3f'})

fig.update_traces(marker=dict(size=8, opacity=0.95))
fig.show()
```
