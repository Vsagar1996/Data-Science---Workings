```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

### Preparing data for Visualization (Pivot table)

```python
# new dataset

df_population = pd.read_csv("population_total.csv")
df_population
```

```python
# dropping null values

df_population.dropna(inplace=True)
```

```python
# making a pivot table - reshape the dataframe and save

df_pivot= df_population.pivot(index="year",columns="country",values="population")
df_pivot
```

```python
# select some countries - simplifying the dataframe and overwrite

df_pivot = df_pivot[["United States","India","China","Indonesia","Brazil"]]
df_pivot
```

```python
df_sample = df_population[df_population["country"].isin(["United States","India","China","Indonesia","Brazil"])]
df_sample
```

# 1)  Static Plots using Pandas

```python
df_pivot.plot(kind="line",xlabel="Year",ylabel="Population",title="Population (1955-2020)",figsize=(8,4))  # figsize(row_size,col_size))
```

```python
# plot using matplotlib

for country in df_sample['country'].unique():
    subset = df_sample[df_sample['country'] == country]
    plt.plot(subset['year'], subset['population'], marker='o', label=country)

plt.title('Population Growth by Country')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```

### 1.3)  Barplot grouped by "n" variables

```python
# selecting a few years (1960,1980,2000,2020)

df_pivot_years = df_pivot[df_pivot.index.isin([1960,1980,2000,2020])]
df_pivot_years
```

```python
# creation of grouped barplot

df_pivot_years.plot(kind="bar",xlabel="year",ylabel="Population",title="Population in selected years",figsize=(8,4))
```

```python
# assigning colours to each bars

df_pivot_years.plot(kind="bar",xlabel="year",ylabel="Population",title="Population in selected years", color=["red","blue","green","brown","yellow"],figsize=(8,4))
```

```python
# horizontal bars

df_pivot_years.plot(kind="barh",xlabel="Population",ylabel="Year",title="Population in selected years",figsize=(8,4))
```

```python
df_sample_years = df_sample[df_sample["year"].isin([1960.0,1980.0,2000.0,2020.0])]
df_sample_years
```

```python
# plot using matplotlib

# dataset
df_sample_years_sort = df_sample_years.sort_values("year")

# Get unique years and countries
years = df_sample_years_sort['year'].unique()
countries = df_sample_years_sort['country'].unique()

bar_width = 0.25
group_spacing = 0.15   # extra space between year groups

# Compute positions with spacing
x = np.arange(len(years)) * (len(countries)*bar_width + group_spacing)

for i, country in enumerate(countries):
    pop = df_sample_years_sort[df_sample_years_sort['country'] == country]['population']
    plt.bar(x + i*bar_width, pop, width=bar_width, label=country)

# Center year labels under groups
plt.xticks(x + (len(countries)-1)*bar_width/2, years)

plt.xlabel("Year")
plt.ylabel("Population (millions)")
plt.title("Population by Country and Year")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# simple example

company=["GOOGL","AMZN","MSFT","FB"]
revenue=[90,136,89,27]
profit=[40,20,34,60]
categories = [revenue,profit]
label = ["revenue","profit"]
bar_width = 0.25
group_spacing = 0.15

x=np.arange(len(company))*(len(categories)*bar_width + group_spacing)

for i, category in enumerate(categories):
    plt.bar(x + i*bar_width, category, width=bar_width, label = label[i])

plt.xticks(x + (len(categories)-1)*bar_width/2, company)
plt.ylabel("Revenue (in USD) & Profit")
plt.xlabel("Companies")
plt.title("Tech Company Stocks")
plt.legend()
```

```python
# horizontal bar plot using matplotlib

# dataset
df_sample_years_sort = df_sample_years.sort_values("year")

# Get unique years and countries
years = df_sample_years_sort['year'].unique()
countries = df_sample_years_sort['country'].unique()

bar_height = 0.25
group_spacing = 0.15   # extra space between year groups

# Compute positions with spacing
y = np.arange(len(years)) * (len(countries)*bar_height + group_spacing)

for i, country in enumerate(countries):
    pop = df_sample_years_sort[df_sample_years_sort['country'] == country]['population']
    plt.barh(y + i*bar_height, pop, height=bar_height, label=country)

# Center year labels under groups
plt.yticks(y + (len(countries)-1)*bar_height/2, years)

plt.ylabel("Year")
plt.xlabel("Population (millions)")
plt.title("Population by Country and Year")
plt.legend()
plt.tight_layout()
plt.show()
```

### 1.6) Histogram using pandas

```python
df_pivot[["United States","Indonesia"]].plot(kind="hist",alpha=0.6)
```

```python
df_pivot[["United States","Indonesia"]].plot(kind="hist",subplots=True)
```

```python
# histogram using matplotlib

# Filter only US and Indonesia
df_US_Indo = df_sample[df_sample["country"].isin(["United States","Indonesia"])]

countries = df_US_Indo["country"].unique()

# Build a list of populations for each country
populations = [df_US_Indo[df_US_Indo["country"] == c]["population"] for c in countries]

# Plot histogram
plt.hist(populations, alpha=0.7, label=countries)

plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Population distribution by countries")
plt.legend()
plt.show()
```

```python
# histogram using matplotlib

df_US_Indo = df_sample[df_sample["country"].isin(["United States","Indonesia"])]

countries = df_US_Indo["country"].unique()

for country in countries:
    subset = df_US_Indo[df_US_Indo['country'] == country]
    plt.hist(subset['population'], alpha=0.6, label=country)

plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Population distribution by countries")
plt.legend()
plt.show()
```
