# Machine Learning with Python

```python
import numpy as np
import pandas as pd
import os
```

```python
# Dataset

# Rooms - no of rooms in the house
# Distance - distance to a major landmar
# Value - value of the house - directly proportional to Rooms and inversely proportional to Distance

df_boston = pd.read_csv('Boston House Prices.csv')
df_boston
```

```python
df_boston.describe()
```

### 1)  Linear Regression with statmodels

General form: y=b0 + b1.x1 + b2.x2 + b3.x3 + ... + bn.xn

y - dependent/outcome/target variable

x1,x2,...,xn - independent/predictor variable

b0 - intercept

b1,b2,...,bn - coefficients

```python
import statsmodels.api as sm
```

### 1.1)  Simple Linear Regression

```python
# define the dependent and independent variable

y = df_boston["Value"]  # dependent variable
x = df_boston["Rooms"]  # independent variable
```

```python
# Exploring the dataset by scatter plots

df_boston.plot(x="Rooms",y="Value",kind="scatter")
```

```python
# Making the regression: Adding a constant and fitting the model

# adding the constant

x = sm.add_constant(x)

# fitting the model

lm = sm.OLS(y,x).fit()  # OLS - ordinary least squares
lm
```

```python
# predict

lm.predict(x)
```

```python
# the regression table - helps to findout how good our fit is

lm.summary()
```

```python
# Linear Regression Equation

# Rooms coeff:	9.1021
# constant coeff: -34.6706

# linear Equation , y=ax+b

y_pred = (9.1021)*x['Rooms'] - 34.6706
```

```python
# Plotting the Regression line

import seaborn as sns
import matplotlib.pyplot as plt

# plotting data points

plt.figure(figsize=(6,4), tight_layout=True)
sns.scatterplot(x=x["Rooms"], y=y)

# plotting the line

sns.lineplot(x=x["Rooms"], y=y_pred, color='red')

# axes

plt.xlim(0)
plt.ylim(0)
plt.savefig('linear_regression')
plt.show()
```

### 1.2) Multiple Linear Regression

```python
# define the dependent and independent variable

y = df_boston["Value"]  # dependent variable
X = df_boston[["Rooms","Distance"]]  # independent variable
```

```python
# Exploring the dataset by scatter plot

ax = df_boston.plot(kind='scatter', x='Rooms', y='Value', color='blue', label='Room vs Value')
df_boston.plot(kind='scatter', x='Distance', y='Value', color='red', label='Distance vs Value', ax=ax)
```

```python
# Making the regression: Adding a constant and fitting the model

X = sm.add_constant(X)  # adding the constant

lm = sm.OLS(y,X).fit()  # fitting the model
lm
```

```python
# Predict values

df_boston['Y_pred'] = lm.predict(X)
df_boston
```

```python
# the regression table - helps to findout how good our fit is

lm.summary()
```

```python
# Multiple Linear Regression Equation

# Rooms coeff:	8.8014
# Distance coeff: 0.4888
# constant coeff: -34.6361

# linear Equation , y = b0 + b1.x1 + b2.x2

y_pred = (8.8014)*X['Rooms'] + (0.4888)*X["Distance"] - 34.6706
y_pred
```

```python
# Scatter plot of actual vs predicted

sns.scatterplot(x=y, y=y_pred, data=df_boston)
plt.xlabel('Actual Y')
plt.ylabel('Predicted Y')
plt.title('Multiple Linear Regression: Actual vs Predicted')
plt.show()
```
