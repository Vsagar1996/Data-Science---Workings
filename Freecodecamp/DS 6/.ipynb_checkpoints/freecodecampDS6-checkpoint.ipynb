# Regular Expressions

```python
# Meta Characters

Metacharacters = """
\d - matches digits (0-9)
\D - matches non digits (0-9)
\w - matches word characters (a-z,A-Z,0-9,'_')
\W - matches non word characters
\s - matches whitespace (space,tab, new line)
\S - matches non whitespace
 . - matches any character except new line
 \ - ignores any special metacharacters - Eg: \.
 ^ - begining of a string  - Eg: ^Hello
 $ - end of a string  - Eg: World$
() - represents group

     Eg: (\d+)-(\d+)-(\d+) matches 876-543-210 in groups (876),(543) and (210)
     
[] - matches characters inside the bracket

     Eg: [789] matches '7','8','9' in '4766819'  - Here [789] means matches '7' or '8' or '9' in the given string
     Eg: [7-9] matches characters from 7 to 9
     Eg: [a-z] matches characters from a to z
     Eg: [A-Z] matches characters from A to Z
     Eg: [a-zA-Z0-9] matches characters from a to z , A to Z and 0 to 9

[^] - matches characters not inside the bracket

     Eg: [^A-Z] matches all characters except from A to Z

| - or

     Eg: 9|8 matches 9 or 8 
     Eg: 9|8\d{2} matches '9' and '876' in '987-876-210'
     Eg: (9|8)\d{2} matches '987' and '876' in '987-876-210'
     Eg: [98]\d{2} matches '987' and '876' in '987-876-210'

\b - matches word boundary
\B - matches all other than word boundary

Note: Word boundaries are staring of the string, ending of the string, between a word and a non word character of a string

|My| |Number| |is| |987|-|768|.| - all '|' symbolically represents word boundaries in 'My Number is 987-768.'

\1 - reference (matches duplicates)

Eg: (123-)\1 matches '123-123' in '123-478-123-123' . Here the first '123-' is not matched as it is not a reference/duplicate.
    (123-)\1\1 dosenot matches any in '123-478-123-123'
    (123-)\1\1 but matches the last '123-' in '123-478-123-123-456-123-658'
    (123-)(-456)\1\2 matches references of 123- and -456 seperately

"""
```

```python
# Quantifiers

Quantifiers= """
* - matches a character 0 or more times (greedy - means matches longest possible)
    Eg: Python!* matches text 'Python!!!!!!' in text 'Python!!!!!!'
    
+ - matches a character 1 or more times (greedy - means matches longest possible)
    Eg: Python!+ matches 'Python!!!!!!' in text 'Python!!!!!!'
    
? - matches a character 0 or 1 instance (lazy - means matches shortest possible)
    Eg: Python!? matches 'Python!' in text 'Python!!!!!'
    
{n} - matches a character exact n instances
    Eg: Python!{3} matches 'Python!!!' in text 'Python!!!!!'

{n,} - matches a character more than n instances
    Eg: Python!{3,} matches 'Python!!!!!' in text 'Python!!!!!'
    
{3,4} - matches a character in the range. ie, min 3 and max 4 instances
    Eg: Python!{3,4} matches 'Python!!!!' in text 'Python!!!!!!!!'

"""
```

```python
# To match string starting with H and ending with l

"""
H.+l - matches 'Hello Worl' in the text 'Hello World' (matches longest possible string)
H.?l - matches 'Hel' in the text 'Hello World' (matches shortest possible string)

"""
```

## 1)  re.search()

```python
text = """Hello World.
I love Python!!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""
```

```python
# return the first match

re.search(r'\d',text) # only matches the first digit
```

```python
# flags

"""
global - don't return after first match
multiline - M - ^ and $ match start/end of line
insensitive - I - case insensitive match
extended - X - ignore whitespace
single line - S - dot matches new line
unicode - match with full unicode
ungreedy - U - make quantifiers lazy
Anchored - A - anchor to start of pattern, or at the end of the most recent search
Jchanged - allow duplicate subpattern names
Dollar end only - $ matches only end of pattern

"""
```

```python
re.search(r'hello',text,flags=re.I)  # here I satnds for case insensitive, neglects the case (lower or upper)
```

### 2)  re.findall()

return all the matches

```python
re.findall(r'\d',text)
```

```python
# match only punctuation

re.findall(r'[^\w\s]',text)
```

```python
# match right date format (MM-DD-YYYY)

text = """
13-04-2021
2021-13-04
2021-04-13
"""

re.findall(r'\d\d-\d\d-\d\d\d\d',text)
```

```python
re.findall(r'\d{2}-\d{2}-\d{4}',text)
```

```python
# match right username format (4-14 characters and only letters or digit)

text = """
username10
abc
10
"""

re.findall(r'[a-zA-Z0-9]{4,14}',text)
```

## 1.3  Dealing with missing data

```python
# to decide whether to delete the columns or delete only the rows having missing data
```

```python
# remove a column or row with .drop, .dropna or .isnull
```

```python
# drop column

df_netflix.drop(["director"], axis=1)
```

```python
# find index of rows having null values in 'director' column

no_director = df_netflix[df_netflix["director"].isnull()].index
no_director
```

```python
# drop rows using only isnull() and not operstor

df_netflix[~(df_netflix['director'].isnull())]
```

```python
# dropna method to drop row - simplest method

df_netflix.dropna(subset=['director'])
```

### filling missing data with fillna() method

Used to replace NaN values by the mean, median (numeric data) or mode

```python
df_netflix["rating"].mode()

# converting the mode to a string - joins the characters

mode_rating = ''.join(df_netflix["rating"].mode())
mode_rating
```

```python
type(mode_rating)
```

```python
# fill the missing values by mode

df_netflix["rating"].fillna(mode_rating,inplace=True)
```

```python
df_netflix["rating"].isnull().sum()
```

```python
df_netflix[df_netflix["rating"].isnull()]
```

```python
# Replace missing values by an arbitrary number with .fillna()

df_netflix["duration"].fillna("0",inplace=True)
```

```python
# backward fill

df_netflix["cast"].fillna(method="bfill",inplace=True)
df_netflix
#df_netflix["cast"].isnull().sum()
```

```python
# forward fill

df_netflix["director"].fillna(method="ffill",inplace=True)
df_netflix
#df_netflix["director"].isnull().sum()
```

## Remove blanks spaces with strip(), lstrip() and rstrip()

```python
# how the strip() method works

movie_title = '  Titanic   '
print(len(movie_title))
# trim leading space with lstrip()
movie_title1 = movie_title.lstrip()
print(len(movie_title1))
# trim trailing space with rstrip()
movie_title2 = movie_title.rstrip()
print(len(movie_title2))
# trim leading and trailing space with strip()
movie_title3 = movie_title.strip()
print(len(movie_title3))
```

```python
# get rid of leading and trailing white spaces with the strip() method

df_movies["title"] = df_movies["title"].str.strip()
#df_movies
```

```python
# get rid of leading and trailing white spaces with the apply method

df_movies["title"] = df_movies["title"].apply(lambda X:X.strip())
#df_movies
```
