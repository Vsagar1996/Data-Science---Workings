The two main data structures in pandas are series (1-D Array) and dataframe (2-D Array, in the form of a table)

The commonly used data structure is dataframe. In dataframe, we have the below terminology

                Row - Row
                Column - Series
                Row values - Observation
                Row heading - Index
                Column name - Features
                Empty cell - NaN (Not a Number)

Note: A single column should only have a single datatype

How to create a dataframe

1. Using Arrays/Numpy Arrays
2. Dictionary (Here key is used as column name and value is used as data)
3. CSV file

## Filter dataframe based on multiple conditions

```python
# find Apple laptops   -  df_laptops["Company"]== "Apple"

# find laptops that cost more than 2000 euros   -  df_laptops["Price_euros"] >= 2000
```

### and - &

```python
# find Apple laptops that cost more than 2000 euros

(df_laptops["Company"]== "Apple") & (df_laptops["Price_euros"] >= 2000)  # brackets are mandatory
```

```python
# filter dataframe based on multiple conditions

df_laptops["Apple and Expensive"] = np.where((df_laptops["Company"]== "Apple") & (df_laptops["Price_euros"] >= 2000), "Both", "Not both")
df_laptops.head()
```

## or - |

```python
# find Apple laptops or laptops that cost more than 2000 euros

(df_laptops["Company"]== "Apple") | (df_laptops["Price_euros"] >= 2000)  # brackets are mandatory
```

```python
# filter dataframe based on either of the two conditions

df_laptops["Apple or Expensive"] = np.where((df_laptops["Company"]== "Apple") | (df_laptops["Price_euros"] >= 2000), "Either", "Neither")
df_laptops.head()
```

```python
# find laptops from Apple or Dell that cost over 2000 euros

df_laptops["(Apple or Dell) but Expensive"] = np.where((((df_laptops["Company"]== "Apple") | (df_laptops["Company"]== "Dell")) & (df_laptops["Price_euros"] >= 2000)), "Satisfied", "Not Satisfied")
df_laptops.head()
```

## Creating a conditional column from more than 2 choices using np.select()

```python
# create an array based on multiple price tiers (more than 2 choices)

# create "conditions" and "values" variables
df_laptops = pd.read_csv('laptop_price2.csv')
conditions = [
    df_laptops["Price_euros"]>3000,
    (df_laptops["Price_euros"]>2000) & (df_laptops["Price_euros"]<=3000),
    (df_laptops["Price_euros"]>1000) & (df_laptops["Price_euros"]<=2000),
    df_laptops["Price_euros"]<1000
]
values = ['Too Expensive','Expensive','Affordable','Cheap']

# set it to a new column

df_laptops['Price Categorization']=np.select(conditions, values,default='unknown')
df_laptops
```

Find the cheapest and most expensive laptops of each companies usig sort_values and duplicated methods



```python
# sort dataframe ascending by company and price - cheapest first and most expensive last

df_laptops = df_laptops.sort_values(["Company","Price_euros"])
df_laptops
```

```python
# check all value categories in "company" column

df_laptops.value_counts("Company")
```

```python
# duplicated values in "Company" column - default - keep='first'

duplicated_first = df_laptops.duplicated('Company',keep='first')
```

```python
# show dataframe with duplicated values in "company" column

df_laptops[duplicated_first]
```

```python
# show dataframe with non duplicated values in "Company" column with lowest prices

df_laptops[~duplicated_first][["Company","Price_euros"]]
```

```python
# keep =False, gives unique (having no duplicates) values in a column

duplicated_false = df_laptops.duplicated('Company',keep=False)
df_laptops[~duplicated_false][["Company","Price_euros"]]
```

### Drop duplicate elements with the .drop_duplicates() method

```python
df_laptops.drop_duplicates(["Company"],keep="last",ignore_index=True)[["Company","Price_euros"]]  # the resulting axis will be labeled as 0,1,2,...
```
