Slicing in numpy

```python
a=np.array([1,2,3,4,5,6])
print(a)
b=a[:4]
print(b)
```

```python
b[0]=100 # change in b and a reflects each other
print(b)
print(a)
```

```python
a[1]=50
print(a)
print(b)
```

```python
a=np.array([1,2,3,4,5,6])
print(a)
b=a[:4].copy()
print(b)
```

```python
b[0]=100
print(b)
print(a)
```

```python
# to find the index of a particular element in an array

A=np.array([1,2,8,9,25,4,9,10])
idx=np.argwhere(A==25)[0][0]
print(idx)
```
