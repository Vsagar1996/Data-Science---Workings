```python
def upper_decor(func):
    def inner():
        res = func()
        return res.upper()
    return inner

@upper_decor
def name():
    return "hello"
print(name())
```

```python
# passing arguments

def upper_decor(func):
    def inner(name):
        res = func(name)
        return res.upper()
    return inner

@upper_decor
def helloname(name):
    return "hello" + " " + name
print(helloname("Jemi"))
```

```python
def upr_dec(func):
    def inr():
        res = func()
        return res.upper()
    return inr
def name():
    return "hello"
f=upr_dec(name)
print(f())
```

#### To Do list

```python
todo_list=[]
def addlist():
    item = input("Enter a new task:")
    todo_list.append(item)
    print(f"{item} added to the todo list")

def displaylist():
    print(".........")
    print("TO Do List:")
    print(".........")
    for index,item in enumerate(todo_list,start=1):
        print(f"{index}-{item}")

def removelist():
    displaylist()
    index = int(input("Enter your item number to remove:"))-1
    if 0<=index<len(todo_list):
        removed_item=todo_list.pop(index)
        print(f"{removed_item} removed from the list")
    else:
        print("Invalid number")

while True:
    print("#########")
    print("To Do List App")
    print("########")
    print("1 - Add to list")
    print("2 - Display list")
    print("3 - Remove from list")
    print("4 - Exit")
    option = input("Select your option:")
    if option == "1":
        addlist()
    elif option == "2":
        displaylist()
    elif option == "3":
        removelist()
    elif option == "4":
        print("Exiting....")
        break
    else:
        print("Invalid option, please try again")
```

```python
# 18)

def mysort(list):
    n=len(list)
    for i in range(0,n):
        for j in range(0,n-i-1):
            if list[j] > list[j+1]:
                list[j],list[j+1] = list[j+1],list[j]
list=[2,4,7,3,9,5]
mysort(list)
print(list)
```
