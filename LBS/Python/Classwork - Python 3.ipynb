Multilevel Inheritance

```python
# Base class

class grandfather:
    def __init__(self,grandfathername):
        self.grandfathername=grandfathername

# intermediate class

class father(grandfather):
    def __init__(self,fathername,grandfathername):
        self.fathername=fathername
        grandfather.__init__(self,grandfathername)

# derived class

class son(father):
    def __init__(self,sonname,fathername,grandfathername):
        self.sonname=sonname
        father.__init__(self,fathername,grandfathername)
        
    def print_family(self):
        print("Grandfather Name: ",self.grandfathername)
        print("Father Name: ",self.fathername)
        print("Son Name: ",self.sonname)

s1=son("ABC","def","123")
s1.print_family()
print(s1.grandfathername)
```

```python
# base class

class school:
    def func1(self):
        print("This function is in school")

# derived class 1 (single inheritance)

class student1(school):
    def func2(self):
        print("This function is in student 1")

# derived class 2 (Another single inheritance)

class student2(school):
    def func3(self):
        print("This function is in student 2")

# derived class 3 (Multiple inheritance)

class student3(student1,school):
    def func4(self):
        print("This function is in student 3")

obj=student3()
obj.func1()
obj.func2()
```

Example for hybrid inheritance

```python
class person:
    def __init__(self,name):
        self.name=name

class employee(person): # Employee inherits from person
    def __init__(self,name,salary):
        super().__init__(name)
        self.salary=salary

# Multiple inheritance

class job:
    def __init__(self,salary):
        self.salary=salary

class employeePersonJob(employee,job): # inherits from both employee and job
    def __init__(self,name,salary):
        employee.__init__(self,name,salary) #initialize employee
        job.__init__(self,salary) # initialize job

# Multilevel Inheritance

class manager(employeePersonJob): # inherits from employeepersonjob
    def __init__(self,name,salary,department):
        employeePersonJob.__init__(self,name,salary) # Explicitly initialize employeeperson job
        self.department=department

# Hierarchical Inheritance

class assistantManager(employeePersonJob): # inherits from employeepersonjob
    def __init__(self,name,salary,team_size):
        employeePersonJob.__init__(self,name,salary) #explicitly initialize employeepersonjob
        self.team_size=team_size

# Hybrid inheritance (Multiple + Multilevel)

class seniorManager(manager,assistantManager): #inherits from both manager and assistantmanager
    def __init__(self,name,salary,department,team_size):
        manager.__init__(self,name,salary,department)
        assistantManager.__init__(self,name,salary,team_size)

emp=employee("John",40000)
print(emp.name,emp.salary)

emp2=employeePersonJob("Alice",50000)
print(emp2.name,emp2.salary)

mgr=manager("Bob",60000,"HR")
print(mgr.name,mgr.salary,mgr.department)

asst_mgr=assistantManager("Charlie", 45000,10)
print(asst_mgr.name,asst_mgr.salary,asst_mgr.team_size)

sen_mgr=seniorManager("David",70000,"Finance",28)
print(sen_mgr.name,sen_mgr.salary,sen_mgr.department,sen_mgr.team_size)          
```

## Abstract Method and Abstract Class

```python
# to find square, squareroot and cube using abstract method

from abc import ABC,abstractmethod

#from abc import *
from math import sqrt

class myclass(ABC):
    @abstractmethod
    def calculate(self,X):
        pass
class sub1(myclass):
    def calculate(self,X):
        print("Square value:",X*X)
class sub2(myclass):
    def calculate(self,X):
        print("Square root",sqrt(X))
class sub3(myclass):
    def calculate(self,X):
        print("Cube value:",X**3)

obj1=sub1()
obj1.calculate(16)
obj2=sub2()
obj2.calculate(16)
obj3=sub3()
obj3.calculate(3)
```

## Encapsulation in Python

### Public, Protected and Private

```python
# Public attributa and method

class Public:
    def __init__(self):
        self.name = "John"  # Public attribute

    def display_name(self):
        print(self.name)  # Public method

obj = Public()
obj.display_name()  # Accessible
print(obj.name)  # Accessible
```

```python
# protected attribute

class Protected:
    def __init__(self):
        self._age = 30  # Protected attribute

class Subclass(Protected):
    def display_age(self):
        print(self._age)  # Accessible in subclass

obj = Subclass()
obj.display_age()
print(obj._age)
```

```python
# Privtae attribute 

class Private:
    def __init__(self):
        self.__salary = 50000  # Private attribute

    def salary(self):
        return self.__salary  # Access through public method

obj = Private()
print(obj.salary())  # Works
# print(obj.__salary)  # Raises AttributeError
```

```python
# Example

class BankAccount:
    def __init__(self):
        self.balance = 1000

    def _show_balance(self):
        print("Balance:", self.balance)  # Protected method

    def __update_balance(self, amount):
        self.balance += amount             # Private method

    def deposit(self, amount):
        if amount > 0:
            self.__update_balance(amount)  # Accessing private method internally
            self._show_balance()           # Accessing protected method
        else:
            print("Invalid deposit amount!")
            
account = BankAccount()
account._show_balance()      # Works, but should be treated as internal
# account.__update_balance(500)  # Error: private method
account.deposit(500)         # Uses both methods internally
```

```python
class Employee:
    def __init__(self):
        self.__salary = 50000  # Private attribute

    def gett_salary(self):    # Getter method
        return self.__salary

    def sett_salary(self, amount):   # Setter method
        if amount > 0:
            self.__salary = amount
        else:
            print("Invalid salary amount!")

emp = Employee()
print(emp.gett_salary())  # Access salary using getter

emp.sett_salary(60000)   # Update salary using setter
print(emp.gett_salary())
```

```python
# to find bank account balance, deposit and withdraw amount

class bankAccount:
    def __init__(self,balance=0):
        self.balance = balance
    def deposit(self):
        amount=float(input("Enter the amount to deposit: "))
        self.balance += amount
    def withdraw(self):
        amount=float(input("Enter the amount to withdraw: "))
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Insufficient balance")
    def display_balance(self):
        print("The balance amount is: ",self.balance)        

myAccountBalance=bankAccount()
myAccountBalance.deposit()
myAccountBalance.withdraw()
myAccountBalance.display_balance()
```

```python
class Student:
    def __init__(self, roll_no, name, marks):
        self.roll_no = roll_no
        self.name = name
        self.marks = marks  # Dictionary of subject: mark
        self.total = sum(marks.values())
        self.average = self.total / len(marks)
        self.grade = self.calculate_grade()

    def calculate_grade(self):
        if self.average>= 90:
            return 'A+'
        elif self.average>= 75:
            return 'A'
        elif self.average>= 60:
            return 'B'
        elif self.average>= 50:
            return 'C'
        else:
            return 'F'

    def display(self):
        marks_str = ' '.join(f"{sub}: {score}" for sub, score in self.marks.items())
        print(f"Roll: {self.roll_no}, Name: {self.name}, Marks: {marks_str}, "
              f"Total: {self.total}, Avg: {self.average:.2f}, Grade: {self.grade}")

class School:
    def __init__(self):
        self.students = []

    def add_student(self, roll_no, name, marks):
        student = Student(roll_no, name, marks)
        self.students.append(student)

    def show_all_students(self):
        print("\nAll Students:\n" + "-"*60)
        for student in self.students:
             student.display()

    def show_rankings(self):
        print("\nStudent Rankings:\n" + "-"*60)
        sorted_students = sorted(self.students, key=lambda x: x.average, reverse=True)
        for rank, student in enumerate(sorted_students, start=1):
             print(f"Rank {rank}: {student.name} (Roll: {student.roll_no}) - Avg: {student.average:.2f}, Grade: {student.grade}")

# Example usage
school = School()

# Adding students manually (you can make it dynamic using input if needed)
school.add_student(101, "Alice", {"Math": 95, "Science": 88, "English": 92})
school.add_student(102, "Bob", {"Math": 75, "Science": 80, "English": 78})
school.add_student(103, "Charlie", {"Math": 60, "Science": 55, "English": 70})
school.add_student(104, "Daisy", {"Math": 45, "Science": 50, "English": 40})

# Display all students
school.show_all_students()

# Show rankings
school.show_rankings()
```
