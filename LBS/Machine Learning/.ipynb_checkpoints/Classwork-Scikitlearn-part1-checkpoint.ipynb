## Probability

```python
# Tossing a coin and getting head on the first throw

fav = 1
total = 2
P_head = fav/total
print("P(getting head)= ",P_head)
```

```python
# rollinga a dice and getting a even number

fav = 3
total = 6
P_even = fav/total
print("P(getting 6)= ",P_even)
```

```python
# complimentary probability- not not getting even

P_not_even = 1-P_even
print("P(not getting even)= ",1-P_even)
```

```python
# probability of A = (getting even) and B = (greater than 4) while rolling a dice

P_A = 3/6
P_B = 2/6
P_A_and_B = 1/6
P_A_or_B = P_A + P_B - P_A_and_B
print("P(A or B)= ",round(P_A_or_B, 3))
```

```python
# In a class of 100 students, 40 like maths, 30 like science and 20 like both. Find probability of P(M|S) = P(M given S is true)

P_M = 40/100
P_S = 30/100
P_M_and_S = 20/100
print("P(M|S)= ",(P_M_and_S)/P_S)
```

# Save & Load Model

Machine Learning with Python: Save and Load trained model

```python
df = pd.DataFrame({"town":["monroe township","monroe township","monroe township","monroe township","monroe township","west windsor","west windsor","west windsor","west windsor","robins ville","robins ville","robins ville","robins ville"],
                          "area":[2600,3000,3200,3600,4000,2600,2800,3300,3600,2600,2900,3100,3600],
                           "price":[550000,565000,610000,680000,725000,585000,615000,650000,710000,575000,600000,620000,695000]
                          })
df
```

```python
# save model to a file using python pickle

import pickle
with open ("model_pickle","wb") as file:  # wb - 'write binary' - saves in binary format
    pickle.dump(model,file) # dump model into file
```

```python
# load saved model

with open ("model_pickle","rb") as file:
    mp = pickle.load(file)

print(mp.coef_)
print(mp.intercept_)
print(mp.predict([[3000]]))
```

```python
# save trained model using joblib

import joblib

joblib.dump(model,"model_joblib")
```

```python
# load saved model using joblink

mj = joblib.load("model_joblib")

print(mj.coef_)
print(mj.intercept_)
print(mj.predict([[5000]]))
```

```python
from sklearn.model_selection import train_test_split
```

Syntax - 

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 42)

X - features (independent variables)

Y - Target (dependent variables)

test_size - proportion of the dataset for testing

train_size - proportion for training (usually 1-(test_size))

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Question 1 - Simple linear regression - house price prediction

X = np.array([[500],[800],[1000],[1200],[1500],[2000]]) # house rsize in sq ft
Y = np.array([2500*x + 150000 for x in X]).reshape(-1,1) # house price in rupees

# create the model
model = LinearRegression()

# train the model
model.fit(X,Y)

# print the slopw and intercept

print("Slope (m):", model.coef_[0][0])
print("intercept (c):", model.intercept_[0])

# predict price for a new house size 1800 sq.ft
new_size = np.array([[1800]])
predicted_price = model.predict(new_size)
print("Predicted price for 1800 sq.ft house:", predicted_price[0][0])

# predict price for a new house size 2500  & 2000 sq.fts
new_size = np.array([[2500],[2000]])
predicted_price = model.predict(new_size)
print("Predicted price for 2500 & 2000 sq.ft houses:", predicted_price[0][0], "&", predicted_price[1][0])
```

```python
# Question 2 - Simple linear regression - Salary prediction

X = np.array([[1],[2],[3],[4],[5],[6]]) # years of experience
Y = np.array([5000*x + 20000 for x in X]).reshape(-1,1) # salaries

# create the model
model = LinearRegression()

# train the model
model.fit(X,Y)

# print the slopw and intercept

print("Slope (m):", model.coef_[0][0])
print("intercept (c):", model.intercept_[0])

# predict salary for 7 years of experience
new_exp = np.array([[7]])
predicted_salary = model.predict(new_exp)
print("Predicted salary for 7 years of experience:", predicted_salary[0][0])

# plot the graph

import matplotlib.pyplot as plt

# plot the points
plt.figure(figsize=(8,5))
plt.scatter(X,Y)

# plot the regression line
plt.plot(X,Y)

# label slope and intercept visually
slope = 5000
intercept = 20000

# show a point to illustrate intercept

plt.scatter(0,intercept, color='black')
plt.text(0,intercept + 1000, "Intercept (c = 20000)", fontsize = 10)

# show slope explanation
plt.text(3, 5000*3 + 20000 + 3000, "Slope (m = 5000)", fontsize = 10)

# graph labels
plt.xlabel("Experience (years)")
plt.ylabel("Salary")
plt.title("Simple Linear Regression: Slope and Intercept")

plt.grid()
plt.show()
```

```python
# another way

data = pd.DataFrame({
    "outlook":["sunny","sunny","overcast","rainy","rainy","rainy","overcast","sunny","sunny","rainy","sunny","overcast","overcast","rainy"],
    "humidity":["high","high","high","high","normal","normal","normal","high","normal","normal","normal","high","normal","high"],
    "windy":["weak","strong","weak","weak","weak","strong","weak","weak","weak","weak","strong","strong","weak","strong"],
    "play":["no","no","yes","yes","yes","no","yes","no","yes","yes","yes","yes","yes","no"]
})
from sklearn.preprocessing import LabelEncoder
outlook = LabelEncoder()
humidity = LabelEncoder()
windy = LabelEncoder()
play = LabelEncoder()

data["outlook"] = outlook.fit_transform(data["outlook"])
data["humidity"] = humidity.fit_transform(data["humidity"])
data["windy"] = windy.fit_transform(data["windy"])
data["play"] = play.fit_transform(data["play"])

feature_cols = ["outlook","humidity","windy"]
X = data[feature_cols] # independent variables
Y = data.play # dependent variable

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=0.8,random_state=32)

from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(criterion="gini")
DTC.fit(X_train, Y_train)

plt.figure(figsize=(12,8))  # set a larger figure size
tree.plot_tree(
    DTC,
    feature_names=feature_cols,       # show feature names instead of numbers
    class_names=["No","Yes"],         # label target classes
    filled=True,                      # color nodes by class
    rounded=True,                     # rounded boxes for clarity
    fontsize=10                       # adjust font size
)
plt.show()
```
