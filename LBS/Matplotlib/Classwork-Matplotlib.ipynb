### Lineplot

```python
x=[1,2,3,4,5,6]
y=[50,51,52,48,47,46]
plt.plot(x,y)
```

```python
#plt.plot(x,y,color="indigo",linewidth=3,linestyle="-")  # linestyle="solid"
#plt.plot(x,y,color="indigo",linewidth=3,linestyle=":") # linestyle="dotted"
#plt.plot(x,y,color="k",linewidth=3,linestyle="--") # linestyle = "dashed"
plt.plot(x,y,color="k",linewidth=3,linestyle="-.") # linestyle = "dashdot"

""" line style

'-'          'solid' line style
'--'         'dashed' line style
'-.'         'dashdot' line style
':'          'dotted' line style

"""

"""  Colors
'b'          blue
'g'          green
'r'          red
'c'          cyan
'm'          magenta
'y'          yellow
'k'          black
'w'          white

"""
```

```python
# to give labels for plot

day=[1,2,3,4,5,6]
temp=[50,51,52,48,47,46]
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Whether")
plt.plot(day,temp,color="indigo",linestyle="dotted")
```

```python
# adding marker and marker size

""" different markers

``'.'``         point marker
``','``         pixel marker
``'o'``         circle marker
``'v'``         triangle_down marker
``'^'``         triangle_up marker
``'<'``         triangle_left marker
``'>'``         triangle_right marker
``'1'``         tri_down marker
``'2'``         tri_up marker
``'3'``         tri_left marker
``'4'``         tri_right marker
``'8'``         octagon marker
``'s'``         square marker
``'p'``         pentagon marker
``'P'``         plus (filled) marker
``'*'``         star marker
``'h'``         hexagon1 marker
``'H'``         hexagon2 marker
``'+'``         plus marker
``'x'``         x marker
``'X'``         x (filled) marker
``'D'``         diamond marker
``'d'``         thin_diamond marker
``'|'``         vline marker
``'_'``         hline marker
"""

day=[1,2,3,4,5,6]
temp=[50,51,52,48,47,46]
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Whether")
plt.plot(day,temp,color="red",marker="*",linestyle="-",markersize=10)
```

```python
# adding (color:marker:linestyle) in different way

day=[1,2,3,4,5,6]
temp=[50,51,52,48,47,46]
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Whether")
plt.plot(day,temp,"g+:")  # "g+:" - (color:marker:linestyle)
```

```python
# selective color - To select RGB color

plt.plot(day,temp,color="#00FF00",marker="+")  # '#00FF00' is one of the RGB colors
#plt.plot(day,temp,color="#00FFFF",marker="+")  # search HTML color codes to get codes for different RGB colors
```

```python
# alpha property to control transparency of line chart

plt.plot(day,temp,"g*--",alpha=0.5)

"""
alpha range is [0,1]
0 - no transparency
0.5 - half of the transparency
1 - maximum transparency
"""
```

### Labels, Legend & Grid

```python
# adding legend and label

day=[1,2,3,4,5,6]
max_temp=[50,51,52,48,47,46]
min_temp=[15,10,17,21,16,8]
avg_temp=[26,27,22,31,28,20]
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Whether")
plt.plot(day,max_temp,label="max")
plt.plot(day,min_temp,label="min")
plt.plot(day,avg_temp,label="average")
# plt.legend() - 'best' is the default value of location
plt.legend(loc="upper left",shadow=True,fontsize="large")

"""
 Location String      Location Code
    ==================   =============
    'best' (Axes only)   0
    'upper right'        1
    'upper left'         2
    'lower left'         3
    'lower right'        4
    'right'              5
    'center left'        6
    'center right'       7
    'lower center'       8
    'upper center'       9
    'center'             10
"""
```

```python
# adding grid

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Whether")
plt.plot(day,max_temp, marker="*",label="max")
plt.plot(day,min_temp,marker="*",label="min")
plt.plot(day,avg_temp,marker="*",label="average")
plt.legend()
plt.grid()
```

## Bar Charts

#### 1. Single Bar

```python
# simple bar chart showing revenues of major tech companies

company=["GOOGL","AMZN","MSFT","FB"]
revenue=[90,136,89,27]
xpos=np.arange(len(company))
plt.bar(xpos,revenue,color=["r","b","g","c"],label="Revenue")
plt.xticks(xpos,company)
plt.ylabel("Revenue (in USD)")
plt.xlabel("Companies")
plt.title("Tech Company Stocks")
#plt.legend()
#plt.grid()  # can also be used
```

### 2) Multiple Bar Plots

```python
# multiple bars showing revenue and profit of tech companies

company=["GOOGL","AMZN","MSFT","FB"]
revenue=[90,136,89,27]
profit=[40,20,34,60]
xpos=np.arange(len(company))
plt.bar(xpos,revenue,color="red",label="Revenue")
plt.bar(xpos,profit,color="green",label="Profit",alpha=0.3)
plt.xticks(xpos,company)
plt.ylabel("Revenue (in USD) & Profit")
plt.xlabel("Companies")
plt.title("Tech Company Stocks")
plt.legend()
#plt.grid()  # can also be used
```

## Histogram

```python
# bins parameter

plt.hist(blood_sugar,rwidth=0.6,bins=4)
```

```python
# histogram showing normal, Pre-diabetic & diabetic patients distribution

"""
80-100 : Normal
100-125 : Pre-DIabetic
125 above : Diabetic
"""

plt.xlabel("Sugar level")
plt.ylabel("Frequency of patients")
plt.title("Blood sugar chart")
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.9,color="y")
```

# Pie Charts

```python
# chart to track montly expenses

exp_vals = [1400,600,300,410,230]
exp_labels = ["Home Rent","Food","Phone/internet","Car","Other activities"]

plt.pie(x = exp_vals,labels = exp_labels)
plt.show()
```

```python
# Draw pie chart as a perfect circle

plt.pie(exp_vals,labels = exp_labels)
plt.axis("equal")
plt.show()
```

```python
# Show percentage for every pie , specify radius to increase the chart size

plt.pie(exp_vals,labels = exp_labels,shadow=True,autopct='%0.2f%%',radius=2) # 0.0 - no decimal, 0.1 - one decimal and so on
plt.axis("equal")
plt.show()
```

```python
# Explode - To project out particular parts

plt.pie(exp_vals,labels = exp_labels,shadow=True,autopct='%0.1f%%',radius=2, explode=[0,0,0.3,0.2,0.8]) # by default anticlockwise direction
plt.axis("equal")
plt.title("Monthly Expenses in Percentage")
plt.legend()
plt.show()
```

```python
# Explode - in clockwise direction (counter clock) & angle properties

plt.pie(exp_vals,labels = exp_labels,shadow=True,autopct='%1.3f%%',radius=2.5, counterclock=True, explode=[0,0.2,0,0.4,0],startangle=90)
plt.axis("equal")
plt.show()
```

## Scatterplot

It is one of the most common ways to visualize relationships b/w two numerical variables

```python
# Example

X=[5,7,8,7,2,17,2,9,4,11]
Y=[99,86,87,88,100,86,103,87,94,78]

plt.scatter(X,Y)
plt.title("Basic Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y values")
plt.show()
```

```python
# customizing a scatter plot

plt.scatter(X,Y,c="red",marker="*",s=100,alpha=0.7)  # s - size of the points
plt.show()
```

### Colored Scatter Plot (Based on value)

```python
X = np.random.rand(50)  # creates 50 random values b/w 0 and 1
Y = np.random.rand(50)
colors = np.random.rand(50)  # creates 50 random color values for the points
plt.scatter(X,Y,c=colors,cmap='viridis',s=100)
#cmap/colormap - other types are plasma, inferno,magma,cividis,coolwarm,jet
#c = colors gives each dots a color based on a numeric value
plt.colorbar()  # show color scale
plt.title("Color mapped scatter plot")
plt.show()
```

### Bubble Chart style Scatter Plot

scatter plot with size and color encoding

```python
sizes = 1000*np.random.rand(50) # creates 50 random sizes for the dots with range from 1 to 100
X = np.random.rand(50)
colors = np.random.rand(50)
plt.scatter(X,Y,s=sizes,c=colors,cmap="plasma", alpha=0.6, edgecolors="r") #edgecolors="r" gives red border around bubbles for contrast
plt.colorbar()
plt.title("Bubble chart style scatter plot")
plt.show()
```
