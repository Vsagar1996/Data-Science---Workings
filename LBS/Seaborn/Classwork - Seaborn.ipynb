```python
import numpy as np
import pandas as pd
import os
```

```python
# import seaborn

import seaborn as sns
```

```python
# Built in datasets

built_in_datasets = sns.get_dataset_names()
print(built_in_datasets)
```

# Tips Dataset

```python
# loading a particular dataset

df_tips = sns.load_dataset('tips')
df_tips
```

#### Correlation and heatmap

```python
# to find correlation between numerical values

corr = df_tips.select_dtypes(['number']).corr()
corr
```

```python
# to get the heatmap

sns.heatmap(corr)
```

#### Joint plot - different graphs in a single graph

```python
sns.jointplot(data=df_tips,x='total_bill',y='tip')
```

#### Pair plot

```python
# gives plots of all possible pairs of values

sns.pairplot(df_tips)
```

#### Count Plot

```python
sns.countplot(x='sex',data=df_tips)
sns.set_style('darkgrid')
```

```python
sns.countplot(x='time',data=df_tips,hue='smoker')
```

#### Barplot

```python
sns.barplot(x='total_bill',y='smoker',data=df_tips,hue='sex')
```

```python
sns.barplot(x='smoker',y='total_bill',data=df_tips,hue='sex')
```

#### Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=df_tips)
```

# Iris Dataset

```python
df_iris = sns.load_dataset('iris')
df_iris
```

#### Violin plot

```python
sns.violinplot(x='species',y='sepal_length',data=df_iris) 

# x - categorical values
# y - continous values
```

#### Strip Plot

```python
sns.stripplot(x='species',y='sepal_length',data=df_iris, jitter=True, hue='petal_width')

# jitter - ensures data points donot overlap
```

#### histplot

```python
sns.set(style='white')
rs=np.random.RandomState(10)
d = rs.normal(size=100)
sns.histplot(d,kde=True,color='b')

# kde - kernal density estimate - to show the curve
```

#### Line Plot

```python
sns.lineplot(y='petal_length',x='petal_width',hue='species', data = df_iris,style='sepal_length')
```

#### lmplot (using for linear regression)

```python
sns.lmplot(x='sepal_length', y='petal_length',data=df_iris)
```
